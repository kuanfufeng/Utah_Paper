```python
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fnst='ModelFit/trim.slst'
```

```python
fig, ax = plt.subplots(3,2,figsize=(14,8))
with open(fnst, 'r') as f:
    for nst, stnm in enumerate(f):
        fn1='ModelFit/code_lineartrend/AIC_csv/AICBIC_'+stnm.strip()+'.csv'
        fn2='ModelFit/code_smoothGW/AIC_csv/AICBIC_'+stnm.strip()+'.csv'
        pd1=pd.read_csv(fn1)
        pd2=pd.read_csv(fn2)
        tmppd=pd.concat([pd1,pd2])

        norm_aic = (tmppd.aic - tmppd.aic.min()) / (tmppd.aic.max() - tmppd.aic.min())
        norm_bic = (tmppd.bic - tmppd.bic.min()) / (tmppd.bic.max() - tmppd.bic.min())

        tmppd['norm_aic']=norm_aic
        tmppd['norm_bic']=norm_bic
        
        if nst==0:
            allpd=tmppd
        else:
            allpd=pd.concat([allpd,tmppd])
    print(allpd.describe())
    print(allpd.modelcase.unique())

    for k, mod in enumerate(allpd.modelcase.unique()):
        print(k,mod)
        aic=allpd.loc[allpd.modelcase == mod, 'aic'].values #- allpd.loc[allpd.modelcase == "soil", 'aic'].values
        bic=allpd.loc[allpd.modelcase == mod, 'bic'].values #- allpd.loc[allpd.modelcase == "soil", 'aic'].values
        color = plt.cm.Accent(k) 
        ax[0,0].plot(aic,bic,label=str(mod),marker='o', ls="", alpha=0.75, color=color, )
        ax[1,0].plot(aic,label=str(mod),    marker='o', ls="", alpha=0.75, color=color, )
        ax[2,0].plot(bic,label=str(mod),    marker='o', ls="", alpha=0.75, color=color, )

        noaic=allpd.loc[allpd.modelcase == mod, 'norm_aic'].values #- allpd.loc[allpd.modelcase == "soil", 'aic'].values
        nobic=allpd.loc[allpd.modelcase == mod, 'norm_bic'].values #- allpd.loc[allpd.modelcase == "soil", 'aic'].values
        color = plt.cm.Accent(k)
        
        ax[0,1].plot(noaic,nobic,label=str(mod),marker='o', ls="", alpha=0.5, color=color, )
        ax[1,1].plot(noaic,label=str(mod),    marker='o', ls="", alpha=0.5, color=color, )
        ax[2,1].plot(nobic,label=str(mod),    marker='o', ls="", alpha=0.5, color=color, )

    ax[0,1].set_title('Model selection by normalized AIC and BIC')
    ax[1,1].set_title('norm(AIC) values over stations')
    ax[2,1].set_title('norm(BIC) values over stations')
    ax[0,0].set_xlabel("AIC value")
    ax[0,0].set_ylabel("BIC value")
    ax[1,0].set_xlabel("Station number")
    ax[2,0].set_xlabel("Station number")
    ax[1,0].set_ylabel("AIC value")
    ax[2,0].set_ylabel("BIC value")
    
    #figname='AB_'+stnm+'_fitdv.png'
    ax[0,0].set_title('Model selection')
    ax[1,0].set_title('AIC values over stations')
    ax[2,0].set_title('BIC values over stations')
    ax[0,0].set_xlabel("AIC value")
    ax[0,0].set_ylabel("BIC value")
    ax[1,0].set_xlabel("Station ")
    ax[2,0].set_xlabel("Station ")
    ax[1,0].set_ylabel("AIC value")
    ax[2,0].set_ylabel("BIC value")
    
    for i in range(3):
        ax[i,0].legend(bbox_to_anchor=(1.05, 1), loc='upper left')
        for j in range(2):
            ax[i,j].grid(True)

    for i in range(1,3):
        for j in range(2):
            ax[i,j].set_xticks(range(len(allpd.stan.unique())))
            ax[i,j].set_xticklabels(allpd.stan.unique())
            plt.setp(ax[i,j].get_xticklabels(), rotation=-90, ha='center')
            
    
    plt.tight_layout()
    

    figname="AICBICoutput_all2.png"
    plt.savefig(figname)
    plt.show()
    f.close()
```
