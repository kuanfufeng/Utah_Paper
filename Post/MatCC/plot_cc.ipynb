```python
import pygmt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fcsv='CC-dv.csv'
uu=pd.read_csv(fcsv, sep=',')
uu_dv = pd.DataFrame(uu)
#print(uu_dv)
```
